## RESPOSTA QUESTÃO 1 ##


$$
\lim_{{x \to 0}} \frac{{\sqrt{x+9} - 3}}{{x}}
$$

Para resolver essa expressão quando x se aproxima de 0, primeiro substituímos x por 0 na expressão:

$$
\frac{{\sqrt{0+9} - 3}}{{0}}
$$

Isso nos dá uma forma indeterminada do tipo $\frac{0}{0}$, o que significa que precisamos simplificar a expressão antes de calcular o limite.

Vamos simplificar a expressão inicial. Para isso, podemos usar a técnica de racionalização multiplicando o numerador e o denominador pelo conjugado da expressão $\sqrt{x+9} - 3$, que é $ sqrt{x+9} + 3 $:

$$
\lim_{{x \to 0}} \frac{{\sqrt{x+9} - 3}}{{x}} \cdot \frac{{\sqrt{x+9} + 3}}{{\sqrt{x+9} + 3}}
$$

Isso nos dá:

$$
\lim_{{x \to 0}} \frac{{(x+9) - 9}}{{x(\sqrt{x+9} + 3)}}
$$

Simplificando ainda mais:

$$
\lim_{{x \to 0}} \frac{{x}}{{x(\sqrt{x+9} + 3)}}
$$

Agora podemos cancelar o fator x do numerador e do denominador:

$$
\lim_{{x \to 0}} \frac{{1}}{{\sqrt{x+9} + 3}}
$$

Substituímos x por 0 novamente:

$$
\frac{{1}}{{\sqrt{0+9} + 3}} = \frac{{1}}{{3+3}} = \frac{{1}}{{6}}
$$

Portanto, o limite da expressão dada é $\frac{1}{6}$ quando x se aproxima de 0.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definindo a variável simbólica e a função
x = sp.symbols('x')
f = (sp.sqrt(x + 9) - 3) / x

# Calculando o limite
limite = sp.limit(f, x, 0)

# Gerando os pontos para o gráfico
x_vals = np.linspace(-1, 1, 400)
y_vals = [(sp.sqrt(x + 9) - 3) / x if x != 0 else limite for x in x_vals]

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$\\frac{\\sqrt{x+9}-3}{x}$')
plt.axhline(y=limite, color='r', linestyle='--', label=f'Linha Limite: {limite}')
plt.xlabel('x')
plt.ylabel('Valor da Função')
plt.title('Gráfico do Limite')
plt.legend()
plt.grid(True)
plt.show()

# Criando a tabela
pontos_esquerda = np.linspace(-0.1, -0.00001, 5)
pontos_direita = np.linspace(0.00001, 0.1, 5)
valores_esquerda = [(sp.sqrt(x + 9) - 3) / x if x != 0 else limite for x in pontos_esquerda]
valores_direita = [(sp.sqrt(x + 9) - 3) / x if x != 0 else limite for x in pontos_direita]

tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

# Mostrando a tabela em LaTeX
display(Math(tabela_latex))
```

## RESPOSTA QUESTÃO 2 ##

Vamos calcular o limite passo a passo:

Começamos com a expressão dada:

$$
\lim_{{x \to \infty}} \frac{{25x + 7}}{{\sqrt{5x^2 + 2}}}
$$

Para resolver essa expressão quando x se aproxima do infinito, podemos simplificar o termo dentro da raiz quadrada dividindo todos os termos por $ x^2 $

$$
\lim_{{x \to \infty}} \frac{{25x + 7}}{{\sqrt{5x^2 + 2}}} \cdot \frac{{\frac{1}{x^2}}}{{\frac{1}{x^2}}}
$$

Isso nos dá:

$$
\lim_{{x \to \infty}} \frac{{25 + \frac{7}{x}}}{{\sqrt{5 + \frac{2}{x^2}}}}
$$

Agora, quando $ x $ se aproxima do infinito, os termos $ \frac{7}{x} $ e $ \frac{2}{x^2} $ tendem a zero, pois estamos dividindo números finitos por números cada vez maiores. Então, podemos simplificar a expressão para:

$$
\frac{{25 + 0}}{{\sqrt{5 + 0}}}
$$

Isso nos dá:

$$
\frac{{25}}{{\sqrt{5}}}
$$

Finalmente, a raiz quadrada de 5 é irracional, então podemos simplificar a raiz quadrada:

$$
\frac{{25}}{{\sqrt{5}}} = \frac{{25 \cdot \sqrt{5}}}{{5}} = 5 \cdot \sqrt{5}
$$

Portanto, o limite da expressão dada é $ 5 \cdot \sqrt{5} $ quando $ x $ se aproxima do infinito.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definindo a variável simbólica e a função
x = sp.symbols('x')
f = (25*x + 7) / sp.sqrt(5*x**2 + 2)

# Calculando o limite
limite = sp.limit(f, x, sp.oo)

# Gerando os pontos para o gráfico
x_vals = np.linspace(1, 100, 500)
y_vals = [(25*x + 7) / np.sqrt(5*x**2 + 2) for x in x_vals]

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$\\frac{25x + 7}{\\sqrt{5x^2 + 2}}$')
plt.axhline(y=limite, color='r', linestyle='--', label=f'Linha Limite: {limite}')
plt.xlabel('x')
plt.ylabel('Valor da Função')
plt.title('Gráfico do Limite')
plt.legend()
plt.grid(True)
plt.show()

# Criando a tabela
pontos_esquerda = np.linspace(100, 1, 5)
pontos_direita = np.linspace(1000, 2000, 5)
valores_esquerda = [(25*x + 7) / np.sqrt(5*x**2 + 2) for x in pontos_esquerda]
valores_direita = [(25*x + 7) / np.sqrt(5*x**2 + 2) for x in pontos_direita]

tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

# Mostrando a tabela em LaTeX
display(Math(tabela_latex))
```

## RESPOSTA QUESTÃO 3 ##

Vamos calcular o limite da função $ f(x) $ conforme definida:

A função é definida como:

$$
f(x) = \begin{cases}
2x + 1 & \text{ se } x \leq 1 \\
x^2 + 1 & \text{ se } x > 1
\end{cases}
$$

Para encontrar o limite de $ f(x) $ quando $ x $ se aproxima de 1, vamos analisar os limites laterais:

1. Para $ x $ se aproximando de 1 por valores menores ou iguais a 1 ($ x \to 1^- $):
   $$
   \lim_{{x \to 1^-}} f(x) = \lim_{{x \to 1^-}} (2x + 1)
   $$
   Substituímos $ x $ por 1 na expressão $ 2x + 1 $:
   $$
   \lim_{{x \to 1^-}} f(x) = 2(1) + 1 = 3
   $$

2. Para $ x $ se aproximando de 1 por valores maiores que 1 ($ x \to 1^+ $):
   $$
   \lim_{{x \to 1^+}} f(x) = \lim_{{x \to 1^+}} (x^2 + 1)
   $$
   Substituímos $ x $ por 1 na expressão $ x^2 + 1 $:
   $$
   \lim_{{x \to 1^+}} f(x) = (1)^2 + 1 = 2
   $$

Como o limite à esquerda é 3 e o limite à direita é 2, e esses limites laterais são diferentes, o limite geral não existe. Em termos matemáticos, podemos dizer que $ \lim_{{x \to 1}} f(x) $ não existe.



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definindo a variável simbólica e a função
x = sp.symbols('x')
f_left = 2*x + 1
f_right = x**2 + 1

# Calculando os limites laterais
limite_esquerda = sp.limit(f_left, x, 1, dir='-')
limite_direita = sp.limit(f_right, x, 1, dir='+')

# Gerando os pontos para o gráfico
x_vals = np.linspace(0.5, 1.5, 500)
y_vals_left = [2*x + 1 if x <= 1 else np.nan for x in x_vals]
y_vals_right = [x**2 + 1 if x > 1 else np.nan for x in x_vals]

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals_left, label='$2x + 1$ se $x \leq 1$')
plt.plot(x_vals, y_vals_right, label='$x^2 + 1$ se $x > 1$')
plt.axvline(x=1, color='r', linestyle='--', label='Limite $x=1$')
plt.scatter([1], [limite_esquerda], color='blue', label=f'Limite à Esquerda: {limite_esquerda}')
plt.scatter([1], [limite_direita], color='green', label=f'Limite à Direita: {limite_direita}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico do Limite')
plt.legend()
plt.grid(True)
plt.show()

# Criando a tabela LaTeX
pontos_esquerda = np.linspace(0.5, 1, 5)
pontos_direita = np.linspace(1, 1.5, 5)
valores_esquerda = [2*x + 1 for x in pontos_esquerda]
valores_direita = [x**2 + 1 for x in pontos_direita]

tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

# Mostrando a tabela em LaTeX
display(Math(tabela_latex))
```

## RESPOSTA QUESTÃO 4 ##

Vamos calcular o limite passo a passo:

A expressão dada é:

$$
\lim_{{x \to -\infty}} (\sqrt{x^2 - x} - x)
$$

Primeiro, observamos que para $ x $ se aproximando de $-\infty$, $ x^2 $ domina a expressão. Então, podemos simplificar $ \sqrt{x^2 - x} $ para $ \sqrt{x^2} $, que é $ |x| $ quando $ x $ é negativo:

$$
\lim_{{x \to -\infty}} (|x| - x)
$$

Como $ x $ se aproxima de $-\infty$, $ |x| $ é igual a $-x$:

$$
\lim_{{x \to -\infty}} (-x - x)
$$

Simplificando:

$$
\lim_{{x \to -\infty}} (-2x)
$$

Quando $ x $ se aproxima de $-\infty$, $ -2x $ se torna $ +\infty $, já que estamos multiplicando um número negativo por um número muito grande negativo.

Portanto, o limite da expressão é $ +\infty $ quando $ x $ se aproxima de $ -\infty $.



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definindo a variável simbólica e a função
x = sp.symbols('x')
f = sp.sqrt(x**2 - x) - x

# Calculando os limites laterais
limite_esquerda = sp.limit(f, x, -sp.oo)
limite_direita = sp.limit(f, x, -sp.oo, dir='-')

# Gerando os pontos para o gráfico
x_vals = np.linspace(-10, -1, 500)
y_vals = [sp.sqrt(x**2 - x) - x for x in x_vals]

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$\sqrt{x^2 - x} - x$')
plt.axhline(y=limite_esquerda, color='r', linestyle='--', label=f'Limite: {limite_esquerda}')
plt.xlabel('x')
plt.ylabel('Valor da Função')
plt.title('Gráfico do Limite')
plt.legend()
plt.grid(True)
plt.show()

# Criando a tabela LaTeX
pontos_esquerda = np.linspace(-10, -1, 5)
pontos_direita = np.linspace(-10, -1, 5)
valores_esquerda = [sp.sqrt(x**2 - x) - x for x in pontos_esquerda]
valores_direita = [sp.sqrt(x**2 - x) - x for x in pontos_direita]

tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

# Mostrando a tabela em LaTeX
display(Math(tabela_latex))
```

## RESPOSTA QUESTÃO 5 ##

Claro, vamos calcular o limite passo a passo:

Dada a expressão:

$$
\lim_{{x \to \infty}} \frac{{4x^2 + 3x - 5}}{{2x^2 - x + 1}}
$$

Primeiro, dividimos todos os termos do numerador e do denominador por $$ x^2 $$, já que estamos lidando com um limite infinito:

$$
\lim_{{x \to \infty}} \frac{{\frac{{4x^2}}{{x^2}} + \frac{{3x}}{{x^2}} - \frac{5}{{x^2}}}}{{\frac{{2x^2}}{{x^2}} - \frac{x}{{x^2}} + \frac{1}{{x^2}}}}
$$

Simplificando cada fração:

$$
\lim_{{x \to \infty}} \frac{{4 + \frac{3}{x} - \frac{5}{{x^2}}}}{{2 - \frac{1}{x} + \frac{1}{{x^2}}}}
$$

Quando $ x $ se aproxima do infinito, os termos $ \frac{3}{x} $ e $ \frac{5}{{x^2}} $ tendem a zero:

$$
\frac{{4 + 0 - 0}}{{2 - 0 + 0}} = \frac{4}{2} = 2
$$

Portanto, o limite da expressão é 2 quando $ x $ se aproxima do infinito.

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definindo a variável simbólica e a função
x = sp.symbols('x')
f = (4*x**2 + 3*x - 5) / (2*x**2 - x + 1)

# Calculando os limites laterais
limite_esquerda = sp.limit(f, x, sp.oo)
limite_direita = sp.limit(f, x, -sp.oo)

# Gerando os pontos para o gráfico
x_vals = np.linspace(-100, 100, 500)
y_vals = [(4*x**2 + 3*x - 5) / (2*x**2 - x + 1) for x in x_vals]

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Limite: $\\frac{4x^2 + 3x - 5}{2x^2 - x + 1}$')
plt.axhline(y=limite_esquerda, color='r', linestyle='--', label=f'Limite à Esquerda: {limite_esquerda}')
plt.axhline(y=limite_direita, color='g', linestyle='--', label=f'Limite à Direita: {limite_direita}')
plt.xlabel('x')
plt.ylabel('Valor da Função')
plt.title('Gráfico do Limite')
plt.legend()
plt.grid(True)
plt.show()

# Criando a tabela LaTeX
pontos_esquerda = [-100, -50, -10, -5, -1]
pontos_direita = [1, 5, 10, 50, 100]
valores_esquerda = [(4*x**2 + 3*x - 5) / (2*x**2 - x + 1) for x in pontos_esquerda]
valores_direita = [(4*x**2 + 3*x - 5) / (2*x**2 - x + 1) for x in pontos_direita]

tabela_latex = "\\begin{array}{|c|c|c|c|}\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\text{Aprox. à Esquerda} & \\text{Valor} & \\text{Aprox. à Direita} & \\text{Valor} \\\\\n"
tabela_latex += "\\hline\n"
for i in range(len(pontos_esquerda)):
    tabela_latex += f"{pontos_esquerda[i]:.5f} & {valores_esquerda[i]:.5f} & {pontos_direita[i]:.5f} & {valores_direita[i]:.5f} \\\\\n"
tabela_latex += "\\hline\n"
tabela_latex += "\\end{array}"

# Mostrando a tabela em LaTeX
display(Math(tabela_latex))
```
